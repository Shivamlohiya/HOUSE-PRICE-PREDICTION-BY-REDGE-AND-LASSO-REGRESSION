```python
import pandas as pd
```

```python
df=pd.read_csv('Housing.csv')
df.head()
```

```python
df = df.select_dtypes(include=['int','float'])
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df[['area','bedrooms','bathrooms','parking']]
y=df[['price']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import *
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=1)
y_test
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_test_predict=lr.predict(x_test)
x_test_predict
```

```python
r2_score(y_test,x_test_predict)
```

```python
x_test_predict_ridge=lr.predict(x_test)
r2_score(y_test,x_test_predict_ridge)
```

```python
lasso_regression=Lasso(alpha=0.1)
lasso_regression.fit(x_train,y_train)
```

```python
x_test_predict_lasso=lasso_regression.predict(x_test)
r2_score(y_test,x_test_predict_lasso)
```
